# Wisconsin Breast Cancer
## Binary classification problem which records clinical measurements of breast cancer tumors.
### Import required libraries

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

from sklearn.neighbors import KNeighborsRegressor
```

### Load Dataset and Analyse it

```python
cancer_data = load_breast_cancer()
cancer_data.keys()
```

### Training & Testing of Breast Cancer Data
#### 1- Split the data into X & y 

```python
X_train, X_test, y_train, y_test = train_test_split(
    cancer_data.data,
    cancer_data.target,
    random_state = 50,
    test_size = 30,
    stratify= cancer_data.target
)
```

#### 2- Train the model through KNN

```python
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(X_train, y_train)
```

#### 3- Predict the model on Test Data

```python
y_pred = knn.predict(X_test)
```

```python
print(y_pred)
```

#### 4- Evaluate that model through metrics

```python
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc}")
```
